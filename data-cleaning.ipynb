```python
df['born_date'] = df['Born'].str.extract(r'(\d+ \w+ \d{4})')
```

```python
df.head(5)
```

```python
df[~df['Born'].str.match(r'(\d+ \w+ \d{4})', na=False) & df['Born'].notna()]
df.head(5)
```

```python
date_pattern = r'(\d+ \w+ \d{4}|\d{4})'
df['born_date'] = df['Born'].str.extract(date_pattern)
df['born_date'] = pd.to_datetime(df['born_date'], format="mixed", errors='coerce')\

df['died_date'] = df['Died'].str.extract(date_pattern)
df['died_date'] = pd.to_datetime(df['died_date'], format="mixed", errors='coerce')
```

```python
df[['Born','born_date', 'Died', 'died_date']].head(5)
```

```python
location_pattern = r'in ([\w\s-]+), ([\w\s-]+) \((\w+)\)'
df['Born'].str.extract(location_pattern)
```

```python
df[['born_city', 'born_region', 'born_country']] = df['Born'].str.extract(location_pattern, expand=True)

pd.set_option('display.max_columns', None)
df[['Born','born_city', 'born_region', 'born_country']].head(5)
```

```python
keep_columns = ['Roles', 'Sex', 'Full name', 'name', 'born_date', 'died_date', 'born_city', 'born_region', 'born_country']

# Normalize column names in the DataFrame by removing extra spaces and converting to lowercase
df.columns = df.columns.str.strip().str.lower()

# Normalize keep_columns by removing extra spaces and converting to lowercase
keep_columns = [col.strip().lower() for col in keep_columns]

# Get the intersection of existing columns after normalization
existing_columns = list(set(df.columns) & set(keep_columns))

# Filter the DataFrame based on the matching columns
df_clean = df[existing_columns]

# Display the first 5 rows of the filtered DataFrame
df_clean.head(5)
```
