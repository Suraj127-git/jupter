```python
import pandas as pd
import random

# Load the CSV files
csv_1 = pd.read_csv("/Users/suraj/Downloads/Marvel Vs DC NEW.csv")
csv_2 = pd.read_csv("/Users/suraj/Downloads/comic_characters.csv")

def random_universe():
    return random.choice(['Marvel', 'DC'])

# Apply the function to create the new 'Universe' column
csv_1['Universe'] = csv_1.apply(lambda row: random_universe(), axis=1)

# Save the updated CSV if needed
csv_1.to_csv("/Users/suraj/Downloads/Marvel_Vs_DC_NEW_with_Universe.csv", index=False)

csv_merge = pd.merge(csv_1, csv_2, on='Universe', how='inner')

csv_merge.head(5)
```
