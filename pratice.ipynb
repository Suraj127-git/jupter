```python
def extract_first_year(year):
    import re
    match = re.search(r'\d{4}', str(year))  # Find a 4-digit year
    return int(match.group()) if match else None

# Apply the function to clean the 'Year' column
csv_1['Year'] = csv_1['Year'].apply(extract_first_year)

# Now you can calculate the median
median_year = csv_1['Year'].median()
csv_1['Year'].fillna(median_year, inplace=True)

csv_1
```

```python
csv_1['Year'] = csv_1['Year'].astype('Int64')
```
